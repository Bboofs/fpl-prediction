```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# path to project directory
path = Path('./')
```

```python
# read in training dataset
train_df = pd.read_csv(path/'data/train_v3.csv', index_col=0, dtype={'season':str})
```

These are the features we get each week direct from fpl and transfermarkt

```python
train_df.info()
```

```python
unique_market_values = (train_df[train_df['season'] < '1920'][['season', 'team', 'relative_market_value_team']]
 .groupby(['season', 'team', 'relative_market_value_team'])).size().reset_index()
```

```python
unique_market_values
```

```python
actual_market_values = unique_market_values.loc[unique_market_values.
                                                groupby(['season', 'team'])[0]
                                                .idxmax()][['season', 'team', 'relative_market_value_team']]
```

```python
actual_market_values
```

```python
train_df.merge(actual_market_values, on=['season', 'team'], how='left', suffixes = ('', '_season'))
```
